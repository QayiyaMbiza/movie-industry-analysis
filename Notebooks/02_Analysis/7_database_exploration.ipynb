# DATABASE EXPLORATION       
The objective of this notebook is to perform an initial exploration of the SQLite database before conducting detailed analysis. This includdes examining the size of the dataset, determining the period by the data, and calculating key descriptive statistics such as average production budget and average box office revenue. 
Understanding these qualities provides context for the analyses performed in the subsequent notebooks. 

```python
import sqlite3
import pandas as pd
```

```python
database = r"C:\Users\bongo\OneDrive\Desktop\GitHub\Movie Analysis\Database\movie_industry.db"

connection = sqlite3.connect(database)

print("Connected successfully.")
```

### COUNT THE MOVIES

```python
# COUNT THE MOVIES

query = """
SELECT COUNT(*) AS Total_Movies
FROM movies;
"""

pd.read_sql(query, connection)
```

### DETERMINE THE YEAR RANGE

```python
# DETERMINE THE YEAR RANGE

query = """
SELECT
    MIN(year) AS First_Year,
    MAX(year) AS Last_Year
FROM movies;
"""

pd.read_sql(query, connection)
```

### AVERAGE BUDGET

```python
# AVERAGE BUDGET

query = """
SELECT
    ROUND(AVG(budget), 2) AS Average_Budget
FROM movies;
"""

pd.read_sql(query, connection)
```

### AVERAGE REVENUE

```python
# AVERAGE REVENUE

query = """
SELECT
    ROUND(AVG(gross),2) AS Average_Revenue
FROM movies;
"""

pd.read_sql(query, connection)
```

### SUMMARY STATISTICS

```python
# SUMMARY STATISTICS

query = """
SELECT
    COUNT(*) AS Total_Movies,
    ROUND(AVG(budget),2) AS Average_Budget,
    ROUND(AVG(gross),2) AS Average_Revenue,
    MIN(year) AS First_Year,
    MAX(year) AS Last_Year
FROM movies;
"""

summary = pd.read_sql(query, connection)

summary
```
