# PRODUCTION ANALYSIS
The objective of this notebook is to analyse the contribution companies to box office success. By examining total revenue, average revenue, and movie production counts, this analysis aims to identify the companies that have consistently produced commercially successful films.

# Research Question

Which production companies have produced the highest-grossing movies?

```python
import sqlite3
import pandas as pd
```

```python
database = r"C:\Users\bongo\OneDrive\Desktop\GitHub\Movie Analysis\Database\movie_industry.db"

connection = sqlite3.connect(database)

print("Connected successfully.")
```

## TOP PRODUCTION COMPANIES BY TOTAL REVENUE

```python
# TOP PRODDUCTION COMPANIES BY TOTAL REVENUE

query = """
SELECT
    company,
    ROUND(SUM(gross), 2) AS Total_Revenue
FROM movies
WHERE company IS NOT NULL
GROUP BY company
ORDER BY Total_Revenue DESC
LIMIT 20;
"""

top_revenue_companies = pd.read_sql(query, connection)

top_revenue_companies
```

## TOP PRODUCTION COMPANIES BY NUMBER OF MOVIES

```python
# TOP PRODUCTION COMPANIES BY NUMBER OF MOVIES

query = """
SELECT
    company,
    COUNT(*) AS Number_of_Movies
FROM movies
WHERE company IS NOT NULL
GROUP BY company
ORDER BY Number_of_Movies DESC
LIMIT 20;
"""

movie_count = pd.read_sql(query, connection)

movie_count
```

# Summary

The analysis demonstrates that production companies differ considerably in both their output and financial performance. Some companies generate high revenue through a large number of movie releases, while others achieve exceptional average revenue despite producing fewer films. Analysing profit alongside revenue provides a more complete measure of commercial success by accounting for production costs. These findings suggest that production company is an important factor associated with box office success.
