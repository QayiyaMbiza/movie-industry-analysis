# POPULARITY ANALYSIS
The objective of this notebook is to investigate whether audience popularity is associated with box office success. This analysis examines the relationship between IMDb ratings, audience vote counts, and box office revenue to determine whether movies receiving higher audience engagement also achieve greater financial performance.


# Research Questions

- Does a movie's popularity score correlate with its financial success?
- Does the number of audience votes correlate with box office revenue?

```python
import sqlite3
import pandas as pd
```

```python
database = r"C:\Users\bongo\OneDrive\Desktop\GitHub\Movie Analysis\Database\movie_industry.db"

connection = sqlite3.connect(database)

print("Connected successfully.")
```

## HIGHEST RATED MOVIES

```python
# HIGHEST RATED MOVIES

query = """
SELECT
    name,
    score,
    gross
FROM movies
WHERE score IS NOT NULL
ORDER BY score DESC
LIMIT 20;
"""

highest_scores = pd.read_sql(query, connection)

highest_scores
```

## MOST VOTED MOVIES

```python
# MOST VOTED MOVIES

query = """
SELECT
    name,
    votes,
    gross
FROM movies
WHERE votes IS NOT NULL
ORDER BY votes DESC
LIMIT 20;
"""

most_voted = pd.read_sql(query, connection)

most_voted
```

## HIGHEST REVENUE MOVIES WITH HIGH SCORES

```python
# HIGHEST REVENUE MOVIES WITH HIGH SCORES

query = """
SELECT
    name,
    score,
    gross
FROM movies
WHERE score >= 8.0
ORDER BY gross DESC
LIMIT 20;
"""

high_score_revenue = pd.read_sql(query, connection)

high_score_revenue
```

## HIGHEST REVENUE MOVIES WITH THE MOST VOTES

```python
# HIGHEST REVENUE MOVIES WITH THE MOST VOTES

query = """
SELECT
    name,
    votes,
    gross
FROM movies
ORDER BY votes DESC, gross DESC
LIMIT 20;
"""

popular_movies = pd.read_sql(query, connection)

popular_movies
```

```python
connection.close()

print("Database connection closed.")
```

# Summary

The SQL analysis suggests that audience popularity is associated with box office performance. Movies receiving higher IMDb ratings and larger numbers of audience votes frequently generate substantial box office revenue, although exceptions exist. Audience engagement appears to be an important contributor to commercial success, but it is unlikely to act independently of other factors such as production budget, genre, release timing, and marketing. These findings reinforce the importance of popularity as one of several factors influencing financial performance in the movie industry.
