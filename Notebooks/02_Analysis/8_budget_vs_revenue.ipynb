# Budget vs. Revenue
The objective of this notebooks is to navigate the relationship a movie's production budget and its box office revenue using SQL queries. By comparing production costs with gross earnings, this analysis aims to determine whether higher budget movies generally achieve greater commercial success and to identify movies that generated exceptional high profits.

# Research Question
Is there a relationship between a movie's production budget and its box office revenue?

```python
import sqlite3
import pandas as pd
```

```python
database = r"C:\Users\bongo\OneDrive\Desktop\GitHub\Movie Analysis\Database\movie_industry.db"

connection = sqlite3.connect(database)

print("Connected successfully.")
```

## TOP 20 HIGHEST BUDGET MOVIES

```python
# TOP 20 HIGHEST BUDGET MOVIES
query = """
SELECT
    name,
    budget,
    gross
FROM movies
ORDER BY budget DESC
LIMIT 20;
"""

highest_budget = pd.read_sql(query, connection)

highest_budget
```

## TOP 20 HIGHEST GROSSING MOVIES

```python
# TOP 20 HIGHEST GROSSING MOVIES

query = """
SELECT
    name,
    budget,
    gross
FROM movies
ORDER BY gross DESC
LIMIT 20;
"""

highest_gross = pd.read_sql(query, connection)

highest_gross
```

## TOP 20 MOST PROFITABLE MOVIES

```python
# TOP 20 MOST PROFITABLE MOVIES

query = """
SELECT
    name,
    budget,
    gross,
    profit
FROM movies
ORDER BY profit DESC
LIMIT 20;
"""

highest_profit = pd.read_sql(query, connection)

highest_profit
```

## HIGHEST MOVIES THAT LOST MONEY

```python
# HIGHEST BUDGET MOVIES THAT LOST MONEY

query = """
SELECT
    name,
    budget,
    gross,
    profit
FROM movies
WHERE profit < 0
ORDER BY budget DESC
LIMIT 20;
"""

losses = pd.read_sql(query, connection)

losses
```

## AVERAGE BUDGET VS. BUDGET REVENUE

```python
# AVERAGE BUDGET VS. BUDGET REVENUE

query = """
SELECT
    ROUND(AVG(budget),2) AS Average_Budget,
    ROUND(AVG(gross),2) AS Average_Revenue,
    ROUND(AVG(profit),2) AS Average_Profit
FROM movies;
"""

summary = pd.read_sql(query, connection)

summary
```

# Summary

The SQL analysis suggests that production budget plays an important role in determining box office performance, as many of the highest-grossing movies also have substantial production budgets. However, the presence of high-budget movies that generated financial losses indicates that budget alone is not sufficient to guarantee commercial success. Other factors, such as genre, popularity, audience reception, and production company, may also contribute significantly to financial performance and will be investigated in the following notebooks.
