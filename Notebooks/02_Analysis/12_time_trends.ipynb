# TIME TRENDS
The objective of this notebook is to examine how box office revenue has changed over time by analysing yearly and decade-based trends. Understanding these trends provides insight into the long-term growth of the movie industry and identifies periods of exceptional commercial performance.

# Research Question

How has movie revenue changed over time?

```python
import sqlite3
import pandas as pd
```

```python
database = r"C:\Users\bongo\OneDrive\Desktop\GitHub\Movie Analysis\Database\movie_industry.db"

connection = sqlite3.connect(database)

print("Connected successfully.")
```

## REVENUE BY YEAR

```python
# REVENUE BY YEAR

query = """
SELECT
    year,
    ROUND(SUM(gross), 2) AS Total_Revenue
FROM movies
GROUP BY year
ORDER BY year;
"""

revenue_year = pd.read_sql(query, connection)

revenue_year
```

## REVENUE BY DECADE

```python
# REVENUE BY DECADE

query = """
SELECT
    year,
    ROUND(SUM(gross), 2) AS Total_Revenue
FROM movies
GROUP BY year
ORDER BY year;
"""

revenue_year = pd.read_sql(query, connection)

revenue_year
```

# Summary

The analysis demonstrates that movie revenue has changed considerably over time. While overall box office revenue generally increases across decades, changes in movie production volume and average revenue per movie suggest that industry growth is influenced by multiple factors. Examining yearly and decade-based trends provides valuable context for understanding how the movie industry has evolved and highlights periods of exceptional commercial performance.
