# RETURN ON INVESTMENT ANALYSIS
The objective of this notebook is to evaluate the financial efficiency of movies by analysing their Return on Investment (ROI). Unlike revenue or profit alone, ROI measures how effectively a movie converted its production budget into financial returns, providing insight into which movies delivered the greatest value for their investment.

# Research Question
Which movies generated the highest return relative to their production budgets?

```python
import sqlite3
import pandas as pd
```

```python
database = r"C:\Users\bongo\OneDrive\Desktop\GitHub\Movie Analysis\Database\movie_industry.db"

connection = sqlite3.connect(database)

print("Connected successfully.")
```

## HIGHEST ROI MOVIES

```python
# HIGHEST ROI MOVIES

query = """
SELECT
    name,
    budget,
    gross,
    ROI
FROM movies
WHERE ROI IS NOT NULL
ORDER BY ROI DESC
LIMIT 20;
"""

highest_roi = pd.read_sql(query, connection)

highest_roi
```

## LOWEST ROI MOVIES

```python
# LOWEST ROI MOVIES

query = """
SELECT
    name,
    budget,
    gross,
    ROI
FROM movies
WHERE ROI IS NOT NULL
ORDER BY ROI ASC
LIMIT 20;
"""

lowest_roi = pd.read_sql(query, connection)

lowest_roi
```

# Summary

The ROI analysis demonstrates that financial efficiency is an important dimension of box office success. While some blockbuster movies generated exceptional profits through large budgets, other films achieved remarkable returns with relatively modest investments. Evaluating ROI alongside revenue and profit provides a more comprehensive understanding of commercial performance and highlights that successful movies are not necessarily those with the highest budgets, but those that maximise returns on investment.
