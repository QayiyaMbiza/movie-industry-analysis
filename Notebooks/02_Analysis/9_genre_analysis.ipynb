### GENRE ANALYSIS
The objective of this notebook is to analyse the financial performance of different movie genres using SQL queries. By comparing average and total box office revenue across genres, this analysis aims to identify which genres have historically achieved the greatest commercial success and determine whether genre is associated with box office performance.

```python
import sqlite3
import pandas as pd
```

```python
database = r"C:\Users\bongo\OneDrive\Desktop\GitHub\Movie Analysis\Database\movie_industry.db"

connection = sqlite3.connect(database)

print("Connected successfully.")
```

```python
## Highest Average by Genre

query = """
SELECT
    genre,
    ROUND(AVG(gross), 2) AS Average_Revenue
FROM movies
GROUP BY genre
ORDER BY Average_Revenue DESC;
"""

average_genre = pd.read_sql(query, connection)

average_genre
```

```python
# Highest Total Revenue by Genre

query = """
SELECT
    genre,
    ROUND(SUM(gross), 2) AS Total_Revenue
FROM movies
GROUP BY genre
ORDER BY Total_Revenue DESC;
"""

total_genre = pd.read_sql(query, connection)

total_genre
```

```python
# Number of movies Produced per Genre

query = """
SELECT
    genre,
    COUNT(*) AS Number_of_Movies
FROM movies
GROUP BY genre
ORDER BY Number_of_Movies DESC;
"""

genre_count = pd.read_sql(query, connection)

genre_count
```

```python
# Top 10 Highest Grossing Movies by Genre

query = """
SELECT
    genre,
    name,
    gross
FROM movies
ORDER BY gross DESC
LIMIT 10;
"""

top_genre_movies = pd.read_sql(query, connection)

top_genre_movies
```

# Summary

The SQL analysis indicates that movie genre is associated with financial performance. Some genres consistently generate higher average and total box office revenue than others, suggesting that audience preferences differ across genres. However, genre performance should also be interpreted alongside production frequency, as genres with more movie releases may naturally accumulate greater total revenue. These findings provide evidence that genre is an important factor to consider when analysing box office success.
