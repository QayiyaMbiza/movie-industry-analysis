# BLOCKBUSTERS
The objective of this notebook is to identify blockbuster movies and analyse the characteristics they share. By examining their revenue, profit, production budgets, audience ratings, and popularity, this analysis aims to determine the common features associated with exceptional box office success.

# Research Question

What characteristics do blockbuster movies have in common?

```python
import sqlite3
import pandas as pd
```

```python
database = r"C:\Users\bongo\OneDrive\Desktop\GitHub\Movie Analysis\Database\movie_industry.db"

connection = sqlite3.connect(database)

print("Connected successfully.")
```

## TOP 20 HIGHEST GROSSING MOVIES

```python
# TOP 20 HIGHEST GROSSING MOVIES

query = """
SELECT
    name,
    genre,
    year,
    budget,
    gross,
    profit
FROM movies
ORDER BY gross DESC
LIMIT 20;
"""

top_grossing = pd.read_sql(query, connection)

top_grossing
```

## HIGHEST PROFIT MOVIES

```python
# HIGHEST PROFIT MOVIES

query = """
SELECT
    name,
    budget,
    gross,
    profit
FROM movies
ORDER BY profit DESC
LIMIT 20;
"""

highest_profit = pd.read_sql(query, connection)

highest_profit
```

# Summary

The analysis indicates that blockbuster movies tend to share several common characteristics, including high production budgets, substantial profits, strong audience engagement, and widespread popularity. While blockbuster films are found across multiple genres, certain genres appear more frequently among the highest-performing movies. These findings suggest that commercial success is influenced by a combination of financial investment, audience reception, and genre rather than a single factor alone.
