# TIME TREND ANALYSIS
The objective of this notebook is to analyse how the movie industry has changed over time. This analysis investigates trends in box office revenue and movie production across different years and decades to identify long-term patterns within the industry.

# Research Questions
- How has total box office revenue changed over time?
- How has movie production changed across decades?

```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.ticker import FuncFormatter
```

```python
clean = r"C:\Users\bongo\OneDrive\Desktop\GitHub\Movie Analysis\Data\Clean\cleaned_movies.csv"

movies = pd.read_csv(clean)

print("Dataset loaded successfully.")

movies.head()
```

```python
def currency_formatter(x, pos):

    if x >= 1_000_000_000:
        return f"${x/1_000_000_000:.0f}B"

    elif x >= 1_000_000:
        return f"${x/1_000_000:.0f}M"

    else:
        return f"${x:,.0f}"
```

# Revenue Trend by Year

This visualisation examines how total box office revenue has changed over time by aggregating the annual revenue of all movies in the dataset.

```python
revenue_by_year = (
    movies
    .groupby("year")["gross"]
    .sum()
)

plt.figure(figsize=(12,6))

plt.plot(
    revenue_by_year.index,
    revenue_by_year.values,
    linewidth=2
)

plt.title("Total Box Office Revenue by Year")

plt.xlabel("Year")
plt.ylabel("Revenue (USD)")

plt.gca().yaxis.set_major_formatter(
    FuncFormatter(currency_formatter)
)

plt.tight_layout()

plt.show()
```

### Findings

The revenue trend line illustrates how total box office revenue has evolved over time. Peaks and declines may reflect changes in audience demand, production investment, or broader economic conditions affecting the film industry.

# Movie Releases by Decade

This visualisation compares the number of movies released during each decade represented in the dataset.

```python
movies_by_decade = (
    movies["decade"]
    .value_counts()
    .sort_index()
)

plt.figure(figsize=(10,6))

plt.bar(
    movies_by_decade.index.astype(str),
    movies_by_decade.values
)

plt.title("Number of Movies Released by Decade")

plt.xlabel("Decade")
plt.ylabel("Number of Movies")

plt.show()
```

### Findings

This chart illustrates changes in movie production across decades. An increase in the number of releases may indicate industry growth, while periods with fewer releases may reflect changes in production capacity or market conditions.

# Summary

The time trends analysis demonstrates how both movie production and box office revenue have evolved throughout the period covered by the dataset. Examining these long-term trends provides valuable context for understanding changes in the movie industry and complements the financial and genre analyses presented in earlier notebooks.
