# UNIVARIATE ANALYSIS
The objective of this notebook is to examine the distribution of the primary numerical variables within the movie dataset. Understanding the distribution of budgets, revenues and IMDb scores provided an overview of the dataset and helps identify skewness, variability and potential outliers before more advanced analysis is performed.

# Research Question:
What are the distributions of movie budgets, revenues and IMDb scores?

```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.ticker import FuncFormatter
```

```python
clean = r"C:\Users\bongo\OneDrive\Desktop\GitHub\Movie Analysis\Data\Clean\cleaned_movies.csv"

movies = pd.read_csv(clean)

print("Dataset loaded successfully.")

movies.head()
```

```python
print(f"Rows: {movies.shape[0]:,}")
print(f"Columns: {movies.shape[1]}")
```

```python
def currency_formatter(x, pos):
    """
    Display large financial values in millions (M) and billions (B).
    """

    if x >= 1_000_000_000:
        return f"${x/1_000_000_000:.0f}B"
    elif x >= 1_000_000:
        return f"${x/1_000_000:.0f}M"
    else:
        return f"${x:,.0f}"
```

# Budget Distribution
The production budget represents the amount invested in producing a movie. Examining its distribution helps identify whether most movies are produced with relatively small budgets or whether high-budget productions dominate the industry.

```python
plt.figure(figsize = (10,6))

plt.hist(
    movies['budget'].dropna(),
    bins=30
)
plt.title('Distribution of Movie Budgets')
plt.xlabel('Budget(USD)')
plt.ylabel('Number of Movies')

plt.gca().xaxis.set_major_formatter(
    FuncFormatter(currency_formatter)
)

plt.show()
```

### Findings

Most movies appear to have relatively modest production budgets, while a smaller number of films have exceptionally large budgets. This indicates a positively skewed distribution, where high-budget productions are less common than lower-budget films.

# Revenue Distribution
Gross revenue represents the total box office earnings generated by a movie. Examining its distribution provides insight into how financial success varies across the industry.

```python
plt.figure(figsize=(10,6))

plt.hist(
    movies["gross"].dropna(),
    bins=30
)

plt.title("Distribution of Box Office Revenue")
plt.xlabel("Revenue (USD)")
plt.ylabel("Number of Movies")

plt.gca().xaxis.set_major_formatter(
    FuncFormatter(currency_formatter)
)

plt.show()
```

### Findings

The distribution of revenue is highly skewed, with most movies generating relatively modest box office earnings while a small number achieve exceptionally high revenues. These extreme values represent blockbuster movies.

# IMDb Distribution 
IMDb scores provide a measure of audience ratings for each movie. Analysing their distribution helps determine whether movies generally receive low, average or high ratings.

```python
plt.figure(figsize=(10,6))

plt.hist(
    movies["score"].dropna(),
    bins=20
)

plt.title("Distribution of IMDb Scores")
plt.xlabel("IMDb Score")
plt.ylabel("Number of Movies")

plt.show()
```

### Findings

Most movies receive average to above-average IMDb ratings, with relatively few movies receiving extremely low or exceptionally high scores. The distribution is concentrated around the middle of the rating scale.

## Summary

The univariate analysis provides an overview of the key numerical variables used throughout the project. Production budgets and box office revenues display positively skewed distributions, indicating that a relatively small number of movies account for the largest financial investments and earnings. IMDb scores are distributed more evenly around the middle of the rating scale, suggesting that most movies receive moderate audience ratings. These findings establish a foundation for the bivariate analyses that follow.
