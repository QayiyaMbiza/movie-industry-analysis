```python
import pandas as pd
```

```python
raw = r"C:\Users\bongo\OneDrive\Desktop\GitHub\Movie Analysis\Data\Raw\movies.csv\movies.csv"

movies = pd.read_csv(raw)

movies.head()
```

```python
movies['profit'] = movies['gross'] - movies['budget']
movies[['name', 'budget', 'gross', 'profit']].head()
```

```python
movies['ROI'] = (
    movies['gross'] - movies['budget']
)/ movies['budget']

movies[['name','ROI']].head()
```

```python
# Remove country information
movies['released'] = movies['released'].str.replace(
    r"\s*\(.*\)",
    "",
    regex=True
)

#Converting to datetime
movies['released'] = pd.to_datetime(
    movies['released'],
    errors = 'coerce'
)

movies[['released']].head()
```

```python
movies['released_month'] = movies['released'].dt.month_name()
movies[['released', 'released_month']].head()
```

```python
movies['release_quarter'] = movies['released'].dt.quarter
movies[['released', 'release_quarter']].head()
```

```python
movies['decade'] = (
    movies['year']//10
)*10
movies[['year', 'decade']].head()
```
