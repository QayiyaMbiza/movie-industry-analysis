## SAVE THE CLEAN DATASET

This notebook defines the location where the cleaned dataset will be stored and exports the processed DataFrame for later use in SQLite and exploratory data analysis.

```python
import os
import pandas as pd
```

```python
# Path to the raw dataset
raw = r"C:\Users\bongo\OneDrive\Desktop\GitHub\Movie Analysis\Data\Raw\movies.csv\movies.csv"

# Loading the dataset
movies = pd.read_csv(raw)

movies.head()
```

```python
# Profit
movies['Profit'] = movies['gross'] - movies['budget']

# Return on Investment (ROI)
movies['ROI'] = (
    movies['gross'] - movies['budget']
) / movies['budget']

# Removing country information from the release date
movies['released'] = movies['released'].str.replace(
    r"\s*\(.*\)",
    "",
    regex=True
)

# Convert to datetime
movies['released_month'] = pd.to_datetime(
    movies['released'],
    errors='coerce'
)

# Release Month
movies['released'] = pd.to_datetime(
    movies['released'],
    errors='coerce'
)
# Release Quarter
movies['release_quarter'] = movies['released'].dt.quarter

# Decade
movies['decade'] = (
    movies['year'] // 10
) * 10
```

```python
clean = r"C:\Users\bongo\OneDrive\Desktop\GitHub\Movie Analysis\Data\Clean"

clean_file = os.path.join(
    clean,
    "cleaned_movies.csv"
)
```

```python
movies.to_csv(
    clean_file,
    index=False
)

print("Clean dataset saved successfully.")

print(f"\nLocation: \n{clean_file}")
```
