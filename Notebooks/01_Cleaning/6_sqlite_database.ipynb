## Loading clean dataset into SQLite

## SQLite Database Connection

The purpose of this notebook is to create a SQLite database and load the cleaned movie dataset into a relationship database management system (RDBMS).
After loading the data, several validation checks are performed to cerify that the dataset has been successfully imported. These include checking the number of rows and columns, inspecting the database schema, and executing sample SQL queries.
The SQLite database will serve as the foundation for the SQL analysis phase of this project.

```python
import sqlite3
import os
import pandas as pd
```

```python
DATABASE_FOLDER = r"C:\Users\bongo\OneDrive\Desktop\GitHub\Movie Analysis\Database"

os.makedirs(
    DATABASE_FOLDER,
    exist_ok=True
)
```

```python
database_path  = os.path.join(
    DATABASE_FOLDER,
    'movie_industry.db'
)
connection = sqlite3.connect(database_path)
print("SQLite database created successfully.")
print(database_path)
print('SQLite database created successfully.')

print(database_path)
```

```python
clean = r"C:\Users\bongo\OneDrive\Desktop\GitHub\Movie Analysis\Data\Clean\cleaned_movies.csv"

movies = pd.read_csv(clean)
movies.head()
```

```python
movies.to_sql(
    name = 'movies',
    con = connection,
    if_exists = 'replace',
    index = False
)
print("Dataset loaded successfully")
```

```python
cursor = connection.cursor()
cursor.execute("""
                SELECT COUNT(*)
                FROM movies;
                """)
row_count = cursor.fetchone()[0]
print(f"Rows loaded into SQLite: {row_count:,}")
```

```python
cursor.execute("""
               PRAGMA table_info(movies);
               
               """)
columns = cursor.fetchall()
print(f"Number of Columns: {len(columns)}")
columns
```

```python
for column in columns:
    print(column[1])
```

## Execute Test Query 1

The first five records are retrieved directly from SQLite.
This confirms that the data can be queried successfully after being loaded into the database. 

```python
query = """
SELECT *
FROM movies
LIMIT 5;
"""

pd.read_sql(query, connection)
```

## Execute Test Query 2
Basic summary statistics are calculated using SQL.
This verifies that aggregate SQL functions operate correctly on the imported dataset.

```python
query = """
SELECT
    COUNT(*) AS TotalMovies,
    AVG(gross) AS AverageRevenue,
    MAX(gross) AS HighestRevenue
FROM movies;

"""
```
