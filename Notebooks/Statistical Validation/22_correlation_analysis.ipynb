# CORRELATION ANALYSIS
The objective of this notebook is to statistically validate the relationships identified during Exploratory Data Analysis (EDA). Pearson's correlation coefficient is used to measure the strength and direction of linear relationships between key numerical variables in the movie dataset. These results provide quantitive evidence to support or challenge the visual patterns observed in previous analyses. 

# Research Questions
- Is there a relationship between a movie's production budget and its box office revenue?
- Is popularity associated with financial success?
- Which movies generated the highest return on investment (ROI)?

```python
import pandas as pd
from scipy.stats import pearsonr
```

```python
clean = (
    r"C:\Users\bongo\OneDrive\Desktop\GitHub"
    r"\Movie Analysis\Data\Clean\cleaned_movies.csv"
)

movies = pd.read_csv(clean)

print("Dataset loaded successfully.")

movies.head()
```

```python
print(f"Rows: {movies.shape[0]:,}")
print(f"Columns: {movies.shape[1]}")

movies.info()
```

```python
def correlation_analysis(data, x, y):
    """
    Calculate and display the Pearson correlation coefficient and p-value for two numerical variables
    """
    correlation, p_value = pearsonr(data[x], data[y])

    print(f'\n{x} vs {y}')
    print(f'Correlation: {correlation:.3f}')
    print(f'P-value: {p_value:.6f}')

    return correlation, p_value
```

## Budget vs. Gross

```python
correlation_analysis(movies, 'budget', 'gross')
```

### Correlation: 0.74
Production budget demonstrated a Strong Positive correlation (r = 0.74) with box office revenue, indication that higher-budget productions generally generate greater financial returns. However, because the relationship is not perfect, production budget alone is insuficient to explain commercial success, suggesting that additional factors also influence revenue generation.

## Gross vs. Votes

```python
correlation_analysis(movies, 'gross', 'votes')
```

### Correlation: 0.631
Audience vote count demonstrates a moderately Strong Positive relationship (r=0.631) with box office revenue. Movies attracting greater audience participation generally achieved higher commercial performance, suggesting that public engagement is more closely associated with financial success than critic ratings alone.

## Gross vs. IMDb Score

```python
correlation_analysis(movies, 'gross', 'score')
```

### Correlation: 0.186
IMDb ratings show only a Very Weak Positive relationship (r=0.186) with box office revenue. While critically well received films may achieve commercial success, audience ratings alone are poor predictors of financial performance.

## Budget vs. ROI

```python
correlation_analysis(movies, 'budget', 'ROI')
```

### Correlation: -0.018
Production budget shows virtually no relationship with return on investment (r= -0.018). This suggests that increasing production expenditure does not necessarily improve investment efficiency. Smaller budget films are capable of generating returns comparable to, or exceeding, those of major blockbuster productions.

## Profit vs. Gross

```python
correlation_analysis(movies, 'Profit', 'gross')
```

### Correlation: 0.984
Gross revenue demonstrates an extremely Strong Positive relationship with profit (r= 0.984). This result is expected because profit is directly derived from box office revenue after accounting for production budget. Consequently, revenue is the strongest predictor of overall profitability within the dataset.
