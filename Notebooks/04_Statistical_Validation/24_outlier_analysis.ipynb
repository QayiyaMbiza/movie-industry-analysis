# Outlier Analysis
The objective of this notebook is to identify and investigate unusually high and low observations within the movie dataset that may influece statistical summaries and exploratory data analysis. By examining extreme values for production budgets, box office revenue, profitability, return on investment (ROI), audience votes, and IMDb ratings, this notebooks provides insight into blockbuster movies, financial failures, and exceptional performers that contribute to the highly skewed distributions identified in previous analyses.

## Research Questions:
- Is there a relationship between production budget and box office revenue?
- Does popularity influence commercial success?
- What qualities define blockbuster movies?
- Which movies achieved the highest financial efficiency (ROI)?

```python
import pandas as pd
```

```python
clean = (
    r"C:\Users\bongo\OneDrive\Desktop\GitHub"
    r"\Movie Analysis\Data\Clean\cleaned_movies.csv"
)

movies = pd.read_csv(clean)

print("Dataset loaded successfully.")

movies.head()
```

```python
print(f"Rows: {movies.shape[0]:,}")
print(f"Columns: {movies.shape[1]}")

movies.info()
```

## Highest Production Budgets
Identify the most expensice movies ever produced within the dataset.

```python
highest_budget = (
    movies[
        ['name', 'year', 'budget', 'gross', 'Profit', 'ROI']
    ]
    .sort_values(
        by='budget',
        ascending=False
    )
    .head(10)
)
highest_budget
```

## Lowest Production Budgets
Identify movies produced with exceptionally small budgets.

```python
lowest_budget = (
    movies[
        ['name', 'year', 'budget', 'gross', 'Profit', 'ROI']
    ]
    .sort_values(
        by='budget',
        ascending=True
    )
    .head(10)
)

lowest_budget
```

## Highest Box Office Revenue
Identify the highest-grossing movies.

```python
highest_revenue = (
    movies[
        ['name', 'year', 'budget', 'gross', 'Profit']
    ]
    .sort_values(
        by='gross',
        ascending=False
    )
    .head(10)
)

highest_revenue
```

## Lowest Box Office Revenue
Identify the poorest performing movies financially.

```python
lowest_revenue = (
    movies[
        ['name', 'year', 'gross', 'budget', 'Profit']
    ]
    .sort_values(
        by='gross',
        ascending=True
    )
    .head(10)
)

lowest_revenue
```

## Highest ROI Movies
Identify the movies that generated the greatest financial return relative to their production budgets.

```python
highest_roi = (
    movies[
        ['name', 'year', 'budget', 'gross', 'Profit', 'ROI']
    ]
    .sort_values(
        by='ROI',
        ascending=False
    )
    .head(10)
)

highest_roi
```

## Lowest ROI Movies
Identify movies that produced the poorest financial returns.

```python
lowest_roi = (
    movies[
        ['name', 'year', 'budget', 'gross', 'Profit', 'ROI']
    ]
    .sort_values(
        by='ROI',
        ascending=True
    )
    .head(10)
)

lowest_roi
```

## Highest Audience Vote Counts
Identify movies with the greatest audience engagement

```python
highest_votes = (
    movies[
        ['name', 'year', 'votes', 'score', 'gross']
    ]
    .sort_values(
        by='votes',
        ascending=False
    )
    .head(10)
)

highest_votes
```

## Highest IMDb Ratings
Identify the highest-rated in the dataset.

```python
highest_scores = (
    movies[
        ['name', 'year', 'score', 'votes', 'gross']
    ]
    .sort_values(
        by=['score', 'votes'],
        ascending=[False, False]
    )
    .head(10)
)

highest_scores
```
