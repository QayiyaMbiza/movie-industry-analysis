# DESCRIPTIVE STATISTICS
The objective of this notebook is to generate descriptive statistical summaries of the movie dataset. These statistics complement the exploratory data analysis by describing the central tendency, variability, distribution and shape of key numerical qualities of movies while helping identify skewed distributions, variability and potential outliers.

## Research Questions
- Budget vs. Revenue
- Genres and Revenue
- Popularity and Financial Success
- Characteristics of Blockbusters
- ROI

```python
import pandas as pd
```

```python
clean = (
    r"C:\Users\bongo\OneDrive\Desktop\GitHub"
    r"\Movie Analysis\Data\Clean\cleaned_movies.csv"
)

movies = pd.read_csv(clean)

print("Dataset loaded successfully.")

movies.head()
```

```python
print(f"Rows: {movies.shape[0]:,}")
print(f"Columns: {movies.shape[1]}")

movies.info()
```

```python
numerical_columns = [
    'budget',
    'gross',
    'Profit',
    'ROI',
    'votes',
    'score',
    'runtime',
    'year'
]

numerical_data = movies[numerical_columns]
numerical_data.head()
```

## Descriptive Statistics

```python
numerical_data.describe().T
```

## Mean

```python
mean = numerical_data.mean()
print(mean)
```

## Median

```python
median = numerical_data.median()
print(median)
```

## Mode

```python
mode = numerical_data.mode()
mode
```

## Minimum and Maximum

```python
summary = pd.DataFrame({
    'Minimum': numerical_data.min(),
    'Maximum': numerical_data.max()
})
summary
```

## Standard Deviation

```python
std = numerical_data.std()
print(std)
```

## Quartiles

```python
quartiles = numerical_data.quantile(
    [0.25, 0.5, 0.75]
)
quartiles
```

## Interquartile Range

```python
iqr = (
    numerical_data.quantile(0.75)
    -
    numerical_data.quantile(0.25)
)
print(iqr)
```

## Variance

```python
variance = numerical_data.var()
print(variance)
```

## Skewness

```python
skewness = numerical_data.skew()
print(skewness)
```

## Kurtosis

```python
kurtosis = numerical_data.kurt()
print(kurtosis)
```
