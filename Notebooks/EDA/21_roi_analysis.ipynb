# RETURN ON INVESTMENT (ROI) ANALYSIS
The objective of this notebook is to evaluate the financial efficiency of movies by analysing their Return on Investment (ROI). Rather than focusing solely on box office revenue, ROI measures how effectively production budgets were converted into profit.

# Research Questions

- Which movies achieved the highest Return on Investment?
- Which movies achieved the lowest Return on Investment?
- How is ROI distributed across the dataset?

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
clean = r"C:\Users\bongo\OneDrive\Desktop\GitHub\Movie Analysis\Data\Clean\cleaned_movies.csv"

movies = pd.read_csv(clean)

print("Dataset loaded successfully.")

movies.head()
```

# ROI Distribution

This visualisation illustrates the distribution of Return on Investment (ROI) across all movies in the dataset.

```python
plt.figure(figsize=(10,6))

plt.hist(
    movies["ROI"].dropna(),
    bins=30
)

plt.title("Distribution of Movie Return on Investment")

plt.xlabel("ROI")

plt.ylabel("Number of Movies")

plt.show()
```

### Findings

The ROI distribution demonstrates the variation in financial performance across movies. While many movies achieve modest returns, a smaller number generate exceptionally high returns on their production budgets.

# Highest Return on Investment Movies

This visualisation identifies the movies that generated the greatest financial return relative to their production budgets.

```python
highest_roi = (
    movies
    .sort_values(
        by="ROI",
        ascending=False
    )
    .head(10)
)

highest_roi
```

```python
plt.figure(figsize=(12,6))

plt.bar(
    highest_roi["name"],
    highest_roi["ROI"]
)

plt.title("Top 10 Movies by Return on Investment")

plt.xlabel("Movie")

plt.ylabel("ROI")

plt.xticks(rotation=45, ha="right")

plt.tight_layout()

plt.show()
```

### Findings

These movies generated the greatest return relative to their production budgets, demonstrating exceptional financial efficiency.

# Lowest Return on Investment Movies

This visualisation identifies the movies that produced the weakest financial returns relative to their production budgets.

```python
lowest_roi = (
    movies
    .sort_values(
        by="ROI",
        ascending=True
    )
    .head(10)
)

lowest_roi
```

```python
plt.figure(figsize=(12,6))

plt.bar(
    lowest_roi["name"],
    lowest_roi["ROI"]
)

plt.title("Lowest 10 Movies by Return on Investment")

plt.xlabel("Movie")

plt.ylabel("ROI")

plt.xticks(rotation=45, ha="right")

plt.tight_layout()

plt.show()
```

### Findings

The lowest ROI movies generated poor financial returns relative to their production budgets. These examples illustrate that large investments do not always translate into commercial success.

# Summary

The ROI analysis evaluated the financial efficiency of movies within the dataset. By comparing the highest and lowest ROI movies alongside the overall ROI distribution, this notebook demonstrates that commercial success is influenced not only by revenue but also by how effectively production budgets are managed.
