# GENRE ANALYSIS
The objective of this notebook is to examine the financial performance and frequency of different movie genres. By comparing total box office revenue and the number of movies produced within each genre, we can identify which genres contributes most significantly to the movie industry.

# Research Question
Which movie genres the highest box office revenue, and which genres occur most frequently within the dataset?


```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.ticker import FuncFormatter
```

```python
clean = r"C:\Users\bongo\OneDrive\Desktop\GitHub\Movie Analysis\Data\Clean\cleaned_movies.csv"

movies = pd.read_csv(clean)

print("Dataset loaded successfully.")

movies.head()
```

```python
def currency_formatter(x, pos):

    if x >= 1_000_000_000:
        return f"${x/1_000_000_000:.0f}B"

    elif x >= 1_000_000:
        return f"${x/1_000_000:.0f}M"

    else:
        return f"${x:,.0f}"
```

# Total Revenue by Genre

This visualisation compares the cumulative box office revenue generated by each movie genre within the dataset.

```python
genre_revenue = (
    movies
    .groupby("genre")["gross"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(10,6))

plt.bar(
    genre_revenue.index,
    genre_revenue.values
)

plt.title("Total Box Office Revenue by Genre")

plt.xlabel("Genre")
plt.ylabel("Revenue (USD)")

plt.xticks(rotation=45)

plt.gca().yaxis.set_major_formatter(
    FuncFormatter(currency_formatter)
)

plt.show()
```

### Findings

This chart shows the total box office revenue generated by each genre. Genres with the highest cumulative revenue have contributed the most financially to the movie industry over the period covered by the dataset.

# Genre Frequency

This visualisation shows how frequently each movie genre appears within the dataset.

```python
genre_frequency = (
    movies["genre"]
    .value_counts()
)

plt.figure(figsize=(10,6))

plt.bar(
    genre_frequency.index,
    genre_frequency.values
)

plt.title("Number of Movies by Genre")

plt.xlabel("Genre")
plt.ylabel("Number of Movies")

plt.xticks(rotation=45)

plt.show()
```

### Findings

The genre frequency chart indicates which genres are most commonly represented within the dataset. Comparing this chart with total revenue provides context when interpreting commercial success across genres.

# Summary

The genre analysis highlights differences in both financial performance and production frequency across movie genres. While some genres generate the greatest total revenue, others appear more frequently within the dataset. Together, these visualisations provide insight into the relationship between genre popularity and commercial success.
