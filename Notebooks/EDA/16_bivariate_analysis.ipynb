# BIVARIATE ANALYSIS
The objective of this notebook is to investigate the relationship between numerical variables within the dataset. By analysing the relationship between production budget, box office revenue, audience votes and IMDb scores, we can identify patterns that may explain commercial success.

```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.ticker import FuncFormatter
```

```python
clean = r"C:\Users\bongo\OneDrive\Desktop\GitHub\Movie Analysis\Data\Clean\cleaned_movies.csv"

movies = pd.read_csv(clean)

print("Dataset loaded successfully.")

movies.head()
```

```python
def currency_formatter(x, pos):

    if x >= 1_000_000_000:
        return f"${x/1_000_000_000:.0f}B"

    elif x >= 1_000_000:
        return f"${x/1_000_000:.0f}M"

    else:
        return f"${x:,.0f}"
```

# Budget vs. Gross Revenue
This scatter plot explores whether movies with larger production budgets generally generate higher box office revenues.

```python
plt.figure(figsize=(10,6))

plt.scatter(
    movies['budget'],
    movies['gross'],
    alpha=0.6
)

plt.title('Budget vs. Box Office Revenue')

plt.xlabel('Budget(USD)')
plt.ylabel('Gross Revenue(USD)')

plt.gca().xaxis.set_major_formatter(
    FuncFormatter(currency_formatter)
)

plt.gca().yaxis.set_major_formatter(
    FuncFormatter(currency_formatter)
)

plt.show()
```

### Findings

The scatter plot illustrates the relationship between production budget and box office revenue. A positive trend would suggest that higher-budget movies generally generate greater revenues, although individual movies may perform differently depending on other factors.

# Gross Revenue vs Audience Votes

This analysis investigates whether movies receiving more audience votes also tend to generate greater box office revenue.

```python
plt.figure(figsize=(10,6))

plt.scatter(
    movies['votes'],
    movies['gross'],
    alpha=0.6
)

plt.title('Gross Revenue vs. Audience Votes')

plt.xlabel('Audience Votes')
plt.ylabel('Gross Revenue(USD)')

plt.gca().yaxis.set_major_formatter(
    FuncFormatter(currency_formatter)
)

plt.show()
```

### Findings

This visualisation examines whether audience engagement is associated with commercial success. A positive relationship may indicate that movies attracting greater public attention also achieve stronger box office performance.

# Gross Revenue vs IMDb Score

This visualisation explores whether higher-rated movies generally earn greater box office revenue.

```python
plt.figure(figsize=(10,6))

plt.scatter(
    movies['score'],
    movies['gross'],
    alpha=0.6
)

plt.title('Gross Revenue vs. IMDb Score')

plt.xlabel('IMDb Score')
plt.ylabel('Gross Revenue(USD)')

plt.gca().yaxis.set_major_formatter(
    FuncFormatter(currency_formatter)
)

plt.show()
```

### Findings

This scatter plot investigates the relationship between audience ratings and commercial success. While higher-rated movies may often perform well financially, the visualisation can also reveal notable exceptions.
