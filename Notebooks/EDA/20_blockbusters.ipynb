The objective of this notebook is to identify the highest-grossing movies within the dataset and explore the characteristics they have in common. By examining commercial blockbusters, this analysis provides insight into the factors that may contribute to exceptional box office success.

```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.ticker import FuncFormatter
```

```python
clean = r"C:\Users\bongo\OneDrive\Desktop\GitHub\Movie Analysis\Data\Clean\cleaned_movies.csv"

movies = pd.read_csv(clean)

print("Dataset loaded successfully.")

movies.head()
```

```python
def currency_formatter(x, pos):

    if x >= 1_000_000_000:
        return f"${x/1_000_000_000:.0f}B"

    elif x >= 1_000_000:
        return f"${x/1_000_000:.0f}M"

    else:
        return f"${x:,.0f}"
```

# Highest Grossing Movies

This analysis identifies the ten highest-grossing movies in the dataset based on worldwide box office revenue.

```python
top_movies = (
    movies
    .sort_values(
        by="gross",
        ascending=False
    )
    .head(10)
)

top_movies
```

```python
plt.figure(figsize=(12,6))

plt.bar(
    top_movies["name"],
    top_movies["gross"]
)

plt.title("Top 10 Highest Grossing Movies")

plt.xlabel("Movie")

plt.ylabel("Gross Revenue (USD)")

plt.xticks(rotation=45, ha="right")

plt.gca().yaxis.set_major_formatter(
    FuncFormatter(currency_formatter)
)

plt.tight_layout()

plt.show()
```

### Findings

The visualisation identifies the highest-grossing movies in the dataset. These movies represent the most commercially successful productions and provide a useful starting point for identifying characteristics associated with blockbuster performance.
